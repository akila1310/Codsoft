```python
fraud_count=dta["Class"].value_counts()
fraud_rate=100*fraud_count/dta.shape[0]
fraud_data=pd.concat([fraud_count,fraud_rate],axis=1).reset_index()
fraud_data.columns=['Class','Count','Percentage']
fraud_data
```

```python
dta_fraud=dta[dta['Class']==1]
dta_not_fraud=dta[dta['Class']==0]
dta_not_fraud_sampled=dta_not_fraud.sample(dta_fraud.shape[0],replace=False,random_state=101)
dta_balanced=pd.concat([dta_not_fraud_sampled,dta_fraud],axis=0).sample(frac=1,replace=False,random_state=101).reset_index().drop('index',axis=1)
dta_balanced
```

```python
fraud_count=dta_balanced["Class"].value_counts()
fraud_rate=100*fraud_count/dta_balanced.shape[0]
fraud_data=pd.concat([fraud_count,fraud_rate],axis=1).reset_index()
fraud_data.columns=['Class','Count','Percentage']
fraud_data
```

```python
X_train,X_test,y_train,y_test=train_test_split(dta_balanced.drop('Class',axis=1),dta_balanced['Class'],test_size=0.2,random_state=101)
print(f'''X_train:{X_train.shape}
X_test:{X_test.shape}
y_train:{y_train.shape}
y_test:{y_test.shape}''')
```

```python
randomForestModel=Pipeline([
    ('Scalar',StandardScaler()),
    ('classifier',RandomForestClassifier())
])
randomForestModel.fit(X_train,y_train)
```

```python
y_pred_logis=randomForestModel.predict(X_test)
y_pred_logis
```

```python
cr=classification_report(y_test,y_pred_logis)
print(cr)
```
