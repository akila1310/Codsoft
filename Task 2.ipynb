```python
k_rng = range(1,10)
sse=[]

for k in k_rng:
    km=KMeans(n_clusters=k)
    km.fit(dta[['petal_length', 'petal_width']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('k_rng')
plt.ylabel("sum of squared errors")
plt.plot(k_rng,sse)
```

```python
km=KMeans(n_clusters=3,random_state=0,)
y_predicted=km.fit_predict(dta[['petal_length','petal_width']])
y_predicted
```

```python
dta['clusters']=y_predicted
dta.head(150)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(dta.species,dta.clusters)
cm
```

```python
true_labels = dta.species
predicted_labels = dta.clusters
cm = confusion_matrix(true_labels,predicted_labels)
class_labels =['setosa','versicolor','virginica']
plt.imshow(cm,interpolation='nearest',cmap=plt.cm.Blues)
plt.title('confusion matrix')
plt.colorbar()
tick_marks=np.arange(len(class_labels))
plt.xticks(tick_marks,class_labels)
plt.yticks(tick_marks,class_labels)

for i in range(len(class_labels)):
    for j in range(len(class_labels)):
        plt.text(j,i,str(cm[i][j]),ha='center',va='center',color='white')
  

```
